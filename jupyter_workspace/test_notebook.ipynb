```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv('AwsDataCatalog_trans_ship_grouping_header_10.csv', header=None)
```

```python
df.head()
```

```python
df.columns = ['order_ref', 'sort_order', 'order_ts']
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df = df.sort_values('order_ts', ascending=False)
```

```python
df.head()
```

```python
df['order_timestamp'] = pd.to_datetime(df['order_ts'], utc=True)
```

```python
df.head()
```

```python
df['local_pst_timestamp'] = pd.to_datetime(df['order_timestamp'], utc=True).dt.tz_convert('US/Pacific')
```

```python
df.head()
```

```python
# Extract components
df['date'] = df['local_pst_timestamp'].dt.date
df['time'] = df['local_pst_timestamp'].dt.time
df['year'] = df['local_pst_timestamp'].dt.year
df['month'] = df['local_pst_timestamp'].dt.month
df['day'] = df['local_pst_timestamp'].dt.day
df['hour'] = df['local_pst_timestamp'].dt.hour
df['minute'] = df['local_pst_timestamp'].dt.minute
df['second'] = df['local_pst_timestamp'].dt.second
df['microsecond'] = df['local_pst_timestamp'].dt.microsecond
```
